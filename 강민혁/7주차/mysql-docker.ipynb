```python
import mysql.connector
from mysql.connector import Error

try:
    connection = mysql.connector.connect(
        host='localhost',
        port=3306,
        user='ga111o',
        password='password',
        database='sample_database'
    )

    if connection.is_connected():
        print("connected!")

        cursor = connection.cursor()
        with open('sample_insert.txt', 'r') as file:
            sample_table_query = file.read()

        # 한번에 넣으니까 애가 정신을 못차려서 한 줄씩 나눠서 넣도록 함
        queries = sample_table_query.split(';')
        for query in queries:
            query = query.strip()
            if query:
                cursor.execute(query)
                connection.commit()

        connection.commit()
        print("done")

except Error as e:
    print(e)

finally:
    if connection.is_connected():
        cursor.close()
        connection.close()
        print("close")
```

```python
import mysql.connector
import pandas as pd


def fetch_all_tables_data(database_name):
    try:
        connection = mysql.connector.connect(
            host='localhost',
            port=3306,
            user='ga111o',
            password='password',
            database=database_name
        )

        cursor = connection.cursor()

        cursor.execute("show tables")
        tables = cursor.fetchall()

        all_tables_data = {}

        for (table_name,) in tables:
            cursor.execute(f"select * from {table_name}")
            results = cursor.fetchall()
            column_names = [i[0] for i in cursor.description]

            df = pd.DataFrame(results, columns=column_names)
            all_tables_data[table_name] = df

        return all_tables_data

    except mysql.connector.Error as err:
        print(err)
        return None
    finally:
        if connection:
            cursor.close()
            connection.close()


all_data = fetch_all_tables_data('sample_database')

if all_data is not None:
    for table_name, df in all_data.items():
        print(f"| {table_name} |")
        print(df)
        print("\n")
```
